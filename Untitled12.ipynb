## **Sales Trend Analysis**
 Trend analysis is to find patterns in data, such as ups & downs. A “trend” is an upwards or downwards shift in a data set over time. Bring out below analysis under Trend Analysis


```python
##profit over a period(monthly)
new_order['Year'] = pd.DatetimeIndex(new_order['Order Date']).year
new_order['Month_Number'] = pd.DatetimeIndex(new_order['Order Date']).month
new_order['Month'] = new_order['Month_Number'].apply(lambda x: calendar.month_abbr[x])

year_month = new_order.groupby(['Year', 'Month','Month_Number']).sum().sort_values(['Year','Month_Number'])
year_month
```

```python
year1=year_month.reset_index()
year1=year1[year1.Year==2018]
year2=year_month.reset_index()
year2=year2[year2.Year==2019]
year1.set_index('Year')
fig,ax=plt.subplots(1,2,figsize=(14,8))
ax[0].set_title("Monthly profit in 2018",size=20)
ax[0].bar(year1['Month'],year1['Profit'])
ax[1].set_title("Monthly profit in 2019",size=20)
ax[1].bar(year2['Month'],year2['Profit'])
ax[0].set_xlabel('month')
ax[0].set_ylabel('profit')
ax[1].set_xlabel('month')
ax[1].set_ylabel('profit')
plt.show()
```

From the above graph we can say there is an profit and loss occured in 2018 but in 2019 there is profits. In 2018 more profits are in the month of October and in 2019 more profits are in the month of January.

# **Sales Target Analysis**
 A sales target is a goal set for a salesperson or sales department measured in revenue or units sold for a specific time.

```python
target = sales.groupby('Category').max()['Target'].reset_index()
actual = orderdetails.groupby('Category').sum().reset_index()
target['Actual_Amount'] =actual['Profit']
target
```

```python
plt.figure(figsize=(10,6))
w=0.4
x_axis=np.arange(3)
plt.bar(x_axis -0.2,target['Target'],w,label='target')
plt.bar(x_axis +0.2,target['Actual_Amount'],w,label='Actual')
plt.xticks(x_axis,target['Category'])
plt.title("Actual vs Target Sales",size=25)
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```

if we consider the sales and ListsofOrder dataset data are in the same year then The above bar graph illustrates the Target and the Actual Amount of Profits per Quantity. None of the category surpass or even meet the target.The most disappointing Category is Furniture, which managed to have a profit of Rs. 2298 with the target of Rs. 11.8K
